## Masking virtual member inheritance by keyword `new`

1.สร้าง console application project

```cmd
dotnet new console --name Lab15_Ex06
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
Derived_2 d2 = new Derived_2();
Base b = (Base) d2;
Derived_1 d1 = (Derived_1) d2;


b.A();
d1.A();
d2.A();

class Base
{
    public virtual void A()
    {
        System.Console.WriteLine("Base.A()");
    }
}
class Derived_1 : Base
{
    public override void A()
    {
        System.Console.WriteLine("Derived_1.A()");
    }
}
class Derived_2 : Derived_1
{
    public new void A()
    {
        System.Console.WriteLine("Derived_2.A()");
    }
}
```

7.อธิบายสิ่งที่พบในการทดลอง

1. การใช้ new: คีย์เวิร์ด new ใช้ในการซ่อน (mask) เมธอดที่สืบทอดมาจากคลาสแม่ ทำให้เมธอดในคลาสลูกไม่ทำงานแบบ Polymorphism ที่ควรจะเป็นเมื่อใช้ override
2. ผลลัพธ์การเรียกเมธอด
- การเรียกเมธอดผ่านตัวแปร Base (เช่น b.A()) จะเรียกเมธอดจากคลาส Base หรือ Derived_1 ขึ้นอยู่กับการประกาศเมธอดในคลาสแม่
- การเรียกเมธอดผ่านตัวแปร Derived_1 (เช่น d1.A()) จะเรียกเมธอดจาก Derived_1
- การเรียกเมธอดผ่านตัวแปร Derived_2 (เช่น d2.A()) จะเรียกเมธอดจาก Derived_2 ที่ถูกซ่อนด้วย new.
3. ไม่มี Polymorphism: เนื่องจากใช้ new เมธอดจะไม่ทำงานตามประเภทของอ็อบเจ็กต์ แต่จะขึ้นอยู่กับชนิดของตัวแปรที่ใช้อ้างอิง
