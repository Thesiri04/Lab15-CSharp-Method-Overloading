2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var animalA = new Dog();
var animalB = new Fish();
var animalC = new Bird();

animalA.Move();
animalB.Move();
animalC.Move();

class Animal
{
   public virtual void Move()
   {
    WriteLine("Move successfully.");
   } 
}

class Dog:Animal
{
    public override void Move()
    {
        WriteLine($"{this.GetType()}: Running on the ground.");
        base.Move();
    }
}
class Fish: Animal
{
    public override void Move()
    {
        WriteLine($"{this.GetType()}: Swimming in the water.");
        base.Move();
    }
}

class Bird: Animal
{
    public override void Move()
    {
        WriteLine($"{this.GetType()}: Flying in the air.");
        base.Move();
    }
}
```

7.อธิบายสิ่งที่พบในการทดลอง

1. การสร้างคลาส Animal
- คลาส Animal มีเมธอด Move() ที่แสดงข้อความ "Move successfully." ซึ่งเป็นเมธอดพื้นฐานที่สามารถถูกใช้ในคลาสลูกๆ ที่สืบทอดมาจาก Animal
2. การใช้ Polymorphism
- คลาส Dog, Fish, และ Bird แต่ละคลาสจะ override เมธอด Move() ของ Animal เพื่อแสดงข้อความที่เหมาะสมกับแต่ละประเภทสัตว์
3. การใช้ base.Move()
- ในแต่ละคลาสลูก (Dog, Fish, Bird) จะเรียก base.Move() เพื่อเรียกใช้งานเมธอด Move() ของคลาส Animal หลังจากที่มีการแสดงข้อความเฉพาะตัวของสัตว์นั้นแล้ว
4. การใช้ WriteLine
- ใช้คำสั่ง WriteLine จาก System.Console เพื่อแสดงข้อความที่เกิดขึ้นในแต่ละเมธอด
